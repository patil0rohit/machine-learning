```python
x = df.drop(columns=['Id', 'SalePrice'])
y = df[['SalePrice']]
```

```python
cat = list(x.columns[x.dtypes=='object'])
con = list(x.columns[x.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                            ('ordinal', OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('num', num_pipe1, con),
                          ('cat', cat_pipe1, cat)]).set_output(transform='pandas')
```

```python
x_pre1 = pre1.fit_transform(x)
x_pre1.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
lr = LinearRegression()
sel = SequentialFeatureSelector(estimator=lr,
                                n_features_to_select='auto',
                                direction='forward')
```

```python
sel.fit_transform(x_pre1, y)
imp_cols = sel.get_feature_names_out()
imp_cols
```

```python
sel_cols = []
for i in imp_cols:
    s = i.split('__')[1]
    sel_cols.append(s)
```

```python
x_sel = x[sel_cols]
x_sel.head()
```

```python
cat_sel = list(x_sel.columns[x_sel.dtypes=='object'])
con_sel = list(x_sel.columns[x_sel.dtypes!='objest'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                            ('ohe', OneHotEncoder(handle_unknown='ignore',
                                                  sparse_output=False))])
```

```python
pre2 =ColumnTransformer([('num', num_pipe2, con_sel),
                         ('cat', cat_pipe2, cat_sel)]).set_output(transform='pandas')
```

```python
x_sel_pre = pre2.fit_transform(x_sel)
x_sel_pre.head()
```
